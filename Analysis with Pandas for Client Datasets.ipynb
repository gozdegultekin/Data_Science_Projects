16. Perform a `merge` using "client_list.csv" and "customer_id_list.csv". Assign the resulting dataframe as `clients`.
17. Perform a `merge` using `clients` and "purchase_log.csv" and limit the subset to only clients who made purchases. Assign the resulting dataframe as `detailed_sales`.
18. Use `groupby` to find the client who spent the most money on purchases. Determine how much he/she spent. HINT: save the intermediate dataframe from using `groupby` as `spenders` before applying slicing to determine the client who spent the most money on purchases.
19. (BONUS) Modify the answer to problem 3 slightly to determine exactly what items where purchased by the top spending client.

```python
client_list =pd.read_csv("client_list.csv")
client_id = pd.read_csv("customer_id_list.csv")
```

```python
pd.merge(client_id, client_list).head()
```

```python
client_id.head()
```

```python
client_list.head()
```

**<font color=blue> The merge operation in pandas is case-sensitive. If two columns contain the same data but their column names are not identical (even due to capitalization differences), pandas treats them as separate columns. As a result, the merged DataFrame may include duplicate columns that represent the same information.**

```python
client_list.columns = client_list.columns.str.upper()
print(client_list.columns)
```

**<font color=blue> Let's do the process again:**

```python
clients = pd.merge(client_id, client_list)
```

```python
clients.head()
```

```python
purchase = pd.read_csv("purchase_log.csv")
purchase.head()
```

```python
detailed_sales=pd.merge(clients, purchase, on= "CUSTOMER_ID", how="inner")

detailed_sales.head()
```

```python
spenders = detailed_sales.groupby("CUSTOMER_ID")["PRICE"].sum()
print(spenders)
```

```python
spenders.max() # max spend
```

```python
spenders.idxmax() # max spending customer id
```

```python
detailed_sales[detailed_sales["CUSTOMER_ID"] == spenders.idxmax()]
# Other information about the relevant customer in the detailed_sales table
```

```python
detailed_sales[detailed_sales["CUSTOMER_ID"] == spenders.idxmax()][["CUSTOMER_ID", "ITEM_DESCRIPTION"]] 
# The customer who spent the most purchased the following items: box of chocolates and bottle of champagne
```
